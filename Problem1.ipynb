```python
import java.util.List;

class Problem1 {
    public static boolean checkList(List<Integer> list) {
        return ((list.get(0) % 2 == 1) && (1 < list.get(0)) && (list.get(0) < 399) && (list.get(0) + 1 == list.get(1)));
    }

    public static int plusCheck(int number) {
        int plusAnswer = 0;

        while (number != 0) {
            plusAnswer += number % 10;
            number /= 10;
        }

        return plusAnswer;
    }

    public static int multiCheck(int number) {
        int multiAnswer = 1;

        while (number != 0) {
            multiAnswer *= number % 10;
            number /= 10;
        }

        return multiAnswer;
    }

    public static int solution(List<Integer> pobi, List<Integer> crong) {
        int answer = Integer.MAX_VALUE;

        int pobiResult = 0;
        int crongResult = 0;

        if (checkList(pobi) && checkList(crong)) {
            int pobiPlus = Math.max(plusCheck(pobi.get(0)), plusCheck(pobi.get(1)));
            int pobiMulti = Math.max(multiCheck(pobi.get(0)), multiCheck(pobi.get(1)));

            pobiResult = Math.max(pobiPlus, pobiMulti);

            int crongPlus = Math.max(plusCheck(crong.get(0)), plusCheck(crong.get(1)));
            int crongMulti = Math.max(multiCheck(crong.get(0)), multiCheck(crong.get(1)));

            crongResult = Math.max(crongPlus, crongMulti);

            if (pobiResult == crongResult) {
                answer = 0;
            }
            else if ((pobiResult > crongResult)) {
                answer = 1;
            }
            else {
                answer = 2;
            }

        }
        else {
            answer = -1;
        }

        return answer;
    }
}
```
